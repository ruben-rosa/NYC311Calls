# The first step in data cleaning is to:
Decide what to do with "bad" data ("UNKNOWN", etc.). Convert to Nan or delete the record.
Make sure all data in a column is in the correct format (convert floats to strings, get rid of the 4 digit extension)
Decide what to do with missing values (NaNs)

```python
# Check Closed Date
df_close_date = df[["Status","Closed Date"]]
print(df_close_date[df_close_date["Closed Date"].isna()])
# Based on the output below, just need to have closed incidents
print("Records to be eliminated : ", df[df["Closed Date"].isna()].count())
df = df[~df["Closed Date"].isna()]

# Dealing with time
# Dates and times are best converted to datetime
# That way they will be useful for analysis because we can compute timedelta objects
dateparser = lambda x:datetime.datetime.strptime(x,'%m/%d/%Y %I:%M:%S %p')
df['Created Date'] = df['Created Date'].apply(lambda x:datetime.datetime.strptime(x,'%m/%d/%Y %I:%M:%S %p'))
df['Closed Date'] = df['Closed Date'].apply(lambda x:datetime.datetime.strptime(x,'%m/%d/%Y %I:%M:%S %p'))
#df['Due Date'] = df['Due Date'].apply(lambda x:datetime.datetime.strptime(x,'%m/%d/%Y %I:%M:%S %p'))
# 
```

```python
# Function to fix the ZIP codes
def fix_zip(input_zip):
    try:
        input_zip = int(float(input_zip))
    except:
        try:
            input_zip = int(input_zip.split('-')[0])
        except:
            return np.NaN
    return str(input_zip)
```

```python
# Testing the fix_zip function
input_zip = '20000-0'
fix_zip(input_zip)
```

```python
# Check Descriptor
print(df["Descriptor"].unique())
df_review = df[df["Descriptor"].isna()]
print("Records to be revised : ", df_review.count())
# After Reviewing the data, decided to change it to "Other (complaint details)"
df["Descriptor"] = df["Descriptor"].fillna("Other (complaint details)")
print(df["Descriptor"].unique())
print("--------------------------------------------------------------------------------------------------")

# Check Location Type
print(df["Location Type"].unique())
df_review = df[df["Location Type"].isna()]
print("Records to be revised : ", df_review.count())
# After Reviewing the data, decided to change it to "Others"
df["Location Type"] = df["Location Type"].fillna("Others")
print(df["Location Type"].unique())
print("--------------------------------------------------------------------------------------------------")

# Check Zip Codes
print(df["Incident Zip"].unique())
df_review = df[df["Incident Zip"].isna()]
print("Records to be revised : ", df_review.count())
# After Reviewing the data, decided to drop the values with zip codes = NaN
df = df[~df["Incident Zip"].isna()]
print(df["Incident Zip"].unique())
print("--------------------------------------------------------------------------------------------------")

# Check City
print(df["City"].unique())
df_review = df[df["City"].isna()]
print("Records to be revised : ", df_review.count())
# After Reviewing the data, decided to change it to "NEW YORK" based on Latitude and Longitude 
df["City"] = df["City"].fillna("NEW YORK")
print(df["City"].unique())
print("--------------------------------------------------------------------------------------------------")

# Address Type
print(df["Address Type"].unique())
df_review = df[df["Address Type"].isna()]
print("Records to be revised : ", df_review.count())
# After Reviewing the data, decided to change it to LATLONG as it has valid data in those fields 
df["Address Type"] = df["Address Type"].fillna("LATLONG")
print(df["Address Type"].unique())
print("--------------------------------------------------------------------------------------------------")

# Regarding Latitude and Longitude missing values, decided to drop all of them as cannot verified specifics locations
print(df["Latitude"].unique())
df_review = df[df["Latitude"].isna()]
print("Records to be revised : ", df_review.count())
# After Reviewing the data, decided to change it to LATLONG as it has valid data in those fields 
df = df[~df["Latitude"].isna()]
print(df["Latitude"].unique())
print("--------------------------------------------------------------------------------------------------")

# Regarding Latitude and Longitude missing values, decided to drop all of them as cannot verified specifics locations
print(df["Longitude"].unique())
df_review = df[df["Longitude"].isna()]
print("Records to be revised : ", df_review.count())
# After Reviewing the data, decided to change it to LATLONG as it has valid data in those fields 
df = df[~df["Longitude"].isna()]
print(df["Longitude"].unique())
print("--------------------------------------------------------------------------------------------------")
```

# Create a column to track the time it takes to close a complaint

```python
df['processing_time'] =  df['Closed Date'] - df['Created Date']
```

# Adding a week and month to track compliants weekly and monthly

```python
df['yyyymm'] = df['Closed Date'].apply(lambda x:datetime.datetime.strftime(x,'%Y%m'))
df['yyyywk'] = df['Closed Date'].apply(lambda x:datetime.datetime.strftime(x,'%Y%w'))
df['day'] = df['Closed Date'].apply(lambda x:datetime.datetime.strftime(x,'%A'))
df['duration_hr'] =df['processing_time'].apply(lambda x:x/np.timedelta64(60,'m'))
df['duration_min'] =df['processing_time'].apply(lambda x:x/np.timedelta64(1,'m'))
```

```python
df['duration_min'] =df['processing_time'].apply(lambda x:x/np.timedelta64(1,'m'))
```

# Group data and plot a bar chart of the incident count

```python
# By Borough
plt.figure(figsize=(16,12))
plt.title("Incidents by Borough")
plot = sns.countplot(x = "Borough",data=df,order=df['Borough'].value_counts(ascending=True).index)
plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
plt.grid(True, which="both", axis="y")
plt.show()
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
borough_type = df.groupby('Borough').size().sort_values(ascending=False)
explode = (0.1, 0, 0, 0, 0)  # only "explode" the 1st slice as it is the larger one
fig1, ax1 = plt.subplots(figsize=(16, 12))
ax1.pie(borough_type.values[:5], explode=explode, labels=borough_type.index[0:5], autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax1.set_title("Pie Chart - Incidents by Boroughs")
plt.show()
```

```python
# Printing the Table of Borough
print(df.groupby(["Borough"]).size().sort_values(ascending=False))
```
